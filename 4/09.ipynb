```python
from google.colab import drive
drive.mount('/content/drive')
import cv2
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
img=cv2.imread('/content/Sunflower_sky_backdrop.jpg')
plt.imshow(img)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.show()
```

```python
# Geometric transformation - Scaling the image and resizing
rows, cols, dim = img.shape

# Create the transformation matrix for scaling
m1 = np.float32([[2, 0, 0], [0, 2, 0]])  # Corrected to (2, 3) shape

# Apply the scaling and resizing transformation
simg = cv2.warpAffine(img, m1, (cols * 3, rows * 2))  # Adjusted target size

# Display the scaled and resized image
plt.imshow(cv2.cvtColor(simg, cv2.COLOR_BGR2RGB))
plt.title("Scaled and Resized Image")
plt.show()

print(simg.shape)
```

```python
# Rotation
angle = 30
m2 = np.float32([[np.cos(np.radians(angle)), -np.sin(np.radians(angle)), 0],
                 [np.sin(np.radians(angle)), np.cos(np.radians(angle)), 0]])

rotated_img = cv2.warpAffine(img, m2, (cols, rows))  # Adjusted target size

# Display the rotated image
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")
plt.show()
```
